```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = np.random.dirichlet([0.1] * 10, 10)
print(s)
plt.barh(range(10), s[:, 0])
plt.barh(range(10), s[:, 1], left=s[:, 0], color='g')
plt.barh(range(10), s[:, 2], left=s[:, 0]+s[:, 1], color='r')
plt.title("Lengths of Strings")
```

```python
np.random.seed(42)
s = np.random.dirichlet([0.1] * 10, 10).transpose()
print(s)
plt.barh(range(10), s[0])
plt.barh(range(10), s[1], left=s[0])
plt.barh(range(10), s[2], left=s[0]+s[1])
plt.barh(range(10), s[3], left=s[0]+s[1]+s[2])
plt.barh(range(10), s[4], left=s[0]+s[1]+s[2]+s[3])
plt.barh(range(10), s[5], left=s[0]+s[1]+s[2]+s[3]+s[4])
plt.barh(range(10), s[6], left=s[0]+s[1]+s[2]+s[3]+s[4]+s[5])
plt.barh(range(10), s[7], left=s[0]+s[1]+s[2]+s[3]+s[4]+s[5]+s[6])
plt.barh(range(10), s[8], left=s[0]+s[1]+s[2]+s[3]+s[4]+s[5]+s[6]+s[7])
plt.barh(range(10), s[9], left=s[0]+s[1]+s[2]+s[3]+s[4]+s[5]+s[6]+s[7]+s[8])
plt.title("Lengths of Strings")
```
